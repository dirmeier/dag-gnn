# DAG-GNN



```r
suppressMessages({
    library(dplyr)
    library(igraph)    
    library(ggraph)
    library(ggplot2)
    library(cowplot)
    
    library(bnlearn)
    library(daggnn)  
})
```

```r
n <- 100L
p <- 10L
```

```r
A <- dagitty::randomDAG(p, 0.25)
```

```r
dag <- bnlearn::random.graph(letters[seq(10)]) %>%
    bnlearn::as.graphNEL() %>%
    igraph::graph_from_graphnel() %>%
    igraph::as_adj() %>%
    as.matrix()
```

```r
A  <- dag * rnorm(p * p, 0, 1)

ordering <- sample(1:p)
A <- A[ordering ,ordering]
colnames(A) <- rownames(A) <- seq(p)
```

```r
igraph::graph_from_adjacency_matrix(A, weighted=TRUE) %>%
  is.dag()
```

```r
Z  <-  matrix(rnorm(n * p, 0, 1), n)
X <- Z %*% solve(diag(p) - A)

head(X)
```

```r
trained.model <- train(X, n_epochs = 300, c = 1, lambda=0)
```

```r
sem(trained.model)
```

```r
plot(trained.model)
```

```r
igraph::graph_from_adjacency_matrix(A, weighted = TRUE) %>%
    ggraph::ggraph(layout = "stress") +
    ggraph::geom_edge_link(
      ggplot2::aes(
        color = weight,
        start_cap = label_rect(node1.name),
        end_cap = label_rect(node2.name)
      ),
      arrow = arrow(length = unit(4, "mm"))
    ) +
    ggraph::geom_node_text(ggplot2::aes(label = name), size = 3) +
    ggraph::scale_edge_color_viridis("Coefficient", end = .8, option = "B") +
    ggraph::scale_edge_width("Direction", range = c(.5, 2), limits = c(.5, 1)) +
    ggraph::theme_graph()
```
